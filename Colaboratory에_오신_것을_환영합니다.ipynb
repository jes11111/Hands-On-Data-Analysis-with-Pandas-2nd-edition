```
weather = pd.read_csv('data/nyc_weather_2018.csv',index_col='date', parse_dates=['date'])
weather.head()
```

```
dp=weather.query('datatype == "PRCP"').groupby(level=0).mean()
dp
```

```
dp.groupby(pd.Grouper(freq='Q')).sum()
```

```
dp.assign(
    quarterly_total_precipitation=lambda x:
    x.groupby(pd.Grouper(freq='Q')).transform(np.sum),
    number_of_days_in_a_month= lambda x:
    x.groupby(pd.Grouper(freq='M')).value.transform('count')
)
```

```
import numpy as np
import pandas as pd

fb = pd.read_csv('data/fb_2018.csv', index_col='date', parse_dates=True).assign(
    trading_volume=lambda x: pd.cut(x.volume, bins=3, labels=['low', 'med', 'high'])
)
fb.head()
```

```
fb.pivot_table(
    index="trading_volume",columns= fb.index.month, values="close",aggfunc='count'
)
```
